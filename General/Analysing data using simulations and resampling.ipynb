## Six roulette wheel spins

```python
# There are 18 red slots, 18 black slots and 2 green slots
slots = ['red']*18 + ['black']*18 + ['green']*2
choice(slots)
```

```python
# Now get 6 spins. There are 2 ways using choice and choices.
# Using choice - here you would need to create a for loop dictating that it needs to do it 6 times
Counter([choice(slots) for i in range(6)])
```

```python
# Using choices - built in function to choose multiple times
Counter(choices(slots, k=6))
```

```python
# Finally, choices also has the ability to use the weights directly. 
# In this way it is a one line code, however can become convoluted if too large.
Counter(choices(['red', 'black', 'green'], [18, 18, 2], k=6))
```

## Deal 20 playing cards without replacement

```python
# One technique is to use Counter. A deck has 16 ten cards and 26 lower cards.
deck = Counter(tens=16, low = 36)
# Choices requires a list to make a decision
deck = list(deck.elements())
# Now simulate the deal
deal = sample(deck, 20)
Counter(deal)
```

```python
# To find out how many cards are still left in the deck (without replacements) we can use a remainder.
deal = sample(deck, 52)  # First need to mention that all cards are dealt 
remainder = deal[20:]  # This asks which cards are left after 20 cards have been dealt
Counter(remainder)
```
